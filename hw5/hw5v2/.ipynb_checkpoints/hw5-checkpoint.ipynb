# HW5

### taking a quick look at data

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] =120
from datetime import datetime
import pandas as pd
import numpy as np
dep2train = pd.read_csv('data/dep2train.csv')
print("Head of training data:")
print(dep2train.head())
dep2test = pd.read_csv('data/dep2test.csv')
print("Head of test data:")
print(dep2test.head())
```

```python
dep2train['date'] = pd.to_datetime(dep2train['date'])
dep2train['fold'] = 'train'
dep2test['date'] = pd.to_datetime(dep2test['date'])
dep2test['fold'] = 'test'

_ = dep2train.plot(y = 'weeklySales', x = 'date', label = 'training data')
_ = dep2test.plot(y = 'weeklySales', x = 'date', label = 'test data')
```

## Q1

```python
woy_train = pd.Series([i.weekofyear for i in dep2train['date']])
woy_test = pd.Series([i.weekofyear for i in dep2test['date']])
```

```python
dep2test['ws_pred'] = [np.mean(dep2train['weeklySales'][woy_train==w]) for w in woy_test]
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred) / y_true))
```

#### (1) the MAPE of the test period

```python
mape = mean_absolute_percentage_error(dep2test['weeklySales'], dep2test['ws_pred'])
print("the MAPE of the test period is {}.".format(mape))
```

#### (2) plot

```python
plt.plot(dep2test['date'], dep2test['ws_pred'])
plt.plot(dep2test['date'], dep2test['weeklySales'])
plt.legend(['predition', 'actual'], loc='upper right')
plt.xlabel('date')
plt.ylabel('weekly sales')
plt.show()
```

## Q2

#### Q2-1

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
train = dep2train.copy()
test = dep2test.copy()
rf_train_x = np.array([train['weeklySales'][i:i+4] for i in range(train.shape[0] - 4)])
rf_train_y = np.array(train['weeklySales'][4:])
```

```python
rfr = RandomForestRegressor(n_estimators=52, random_state=0).fit(rf_train_x, rf_train_y)
```

#### (1) MAPE

```python
data = pd.concat([train['weeklySales'][-4:], test['weeklySales']], ignore_index=True)
rf_test_x = np.array([data[i:i+4] for i in range(data.shape[0] - 4)])
rf_test_y = np.array(data[4:])
rfr_pred = rfr.predict(rf_test_x)
rfr_mape = mean_absolute_percentage_error(rf_test_y, rfr_pred)
print("the MAPE of the test period with previous 4 periods using Random Forest Regressor is {}.".format(rfr_mape))
```

#### (2) plot

```python
plt.plot(dep2test['date'], rfr_pred)
plt.plot(dep2test['date'], dep2test['weeklySales'])
plt.legend(['predition', 'actual'], loc='upper right')
plt.xlabel('date')
plt.ylabel('weekly sales')
plt.show()
```

#### Q2-2

```python
rf_train_x2 = np.array([np.append(train['weeklySales'][i+48:i+52], train['weeklySales'][i]) for i in range(train.shape[0] - 52)])
rf_train_y2 = np.array(train['weeklySales'][52:])
```

```python
rfr2 = RandomForestRegressor(n_estimators=52, random_state=0).fit(rf_train_x2, rf_train_y2)
```

#### (1) MAPE

```python
data2 = pd.concat([train['weeklySales'][-52:], test['weeklySales']], ignore_index=True)
rf_test_x2 = np.array([np.append(data2[i+48:i+52], data2[i]) for i in range(data2.shape[0] - 52)])
rf_test_y2 = np.array(data2[52:])
rfr_pred2 = rfr2.predict(rf_test_x2)
rfr_mape2 = mean_absolute_percentage_error(rf_test_y2, rfr_pred2)
print("the MAPE of the test period with previous 4 periods and same period \nlast year using Random Forest Regressor is {}.".format(rfr_mape2))
```

#### (2) plot

```python
plt.plot(dep2test['date'], rfr_pred2)
plt.plot(dep2test['date'], dep2test['weeklySales'])
plt.legend(['predition', 'actual'], loc='upper right')
plt.xlabel('date')
plt.ylabel('weekly sales')
plt.show()
```

## Q3

#### Q3-1

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ExpSineSquared, WhiteKernel
kernels = RBF() + ExpSineSquared() + WhiteKernel()
```

```python
train3 = dep2train.copy()
test3 = dep2test.copy()
mean = np.mean(train3['weeklySales'])
std = np.std(train3['weeklySales'])
train_y = (train3['weeklySales'] - mean) / std
test_y = (test3['weeklySales'] - mean) / std
train_x = np.array(woy_train).reshape(-1, 1)
test_x = np.array(woy_test).reshape(-1, 1)
```

#### (1) MAPE

```python
gpr = GaussianProcessRegressor(kernel = kernels, random_state=0).fit(train_x, train_y)
gpr_pred = gpr.predict(test_x)
gpr_pred = gpr_pred * std + mean
gpr_mape = mean_absolute_percentage_error(test3['weeklySales'], gpr_pred)
print("the MAPE of the test period using Gaussian Process Regressor with standardized data is {}.".format(gpr_mape))
```

#### (2) plot

```python
plt.plot(dep2test['date'], gpr_pred)
plt.plot(dep2test['date'], dep2test['weeklySales'])
plt.legend(['predition', 'actual'], loc='upper right')
plt.xlabel('date')
plt.ylabel('weekly sales')
plt.show()
```

#### Q3-2

#### (1) MAPE

```python
train_y_o = train3['weeklySales']
train_x_o = train_x
test_y_o = test3['weeklySales']
test_x_o = test_x
gpr_o = GaussianProcessRegressor(kernel=kernels, random_state=0).fit(train_x_o, train_y_o)
gpr_pred_o = gpr_o.predict(test_x_o)
gpr_mape_o = mean_absolute_percentage_error(test3['weeklySales'], gpr_pred_o)
print("the MAPE of the test period using Gaussian Process Regressor is {}.".format(gpr_mape_o))
```

#### (2) plot

```python
plt.plot(dep2test['date'], gpr_pred_o)
plt.plot(dep2test['date'], dep2test['weeklySales'])
plt.legend(['predition', 'actual'], loc='upper right')
plt.xlabel('date')
plt.ylabel('weekly sales')
plt.show()
```

#### Q3-3

```python
train_y = (train3['weeklySales'] - mean) / std
test_y = (test3['weeklySales'] - mean) / std
train_x = np.array(woy_train).reshape(-1, 1)
test_x = np.array(woy_test).reshape(-1, 1)
gpr2 = GaussianProcessRegressor(kernel=kernels, random_state=0).fit(train_x, train_y)
gpr2_preds = []
gpr2_pred = gpr2.predict(test_x[0].reshape(-1,1))
gpr2_pred = gpr2_pred * std + mean
gpr2_preds.append(gpr2_pred[0])
for i in range(1, test_x.shape[0]):
    train_x = np.append(train_x, test_x[i-1])
    train_y = np.append(train_y, test_y[i-1])
    train_x = train_x.reshape(-1, 1)
    test_xi = test_x[i].reshape(-1, 1)
    gpr2.fit(train_x, train_y)
    gpr2_pred = gpr2.predict(test_xi)
    gpr2_pred = gpr2_pred * std + mean
    gpr2_preds.append(gpr2_pred[0])

gpr2_mape = mean_absolute_percentage_error(test3['weeklySales'], np.array(gpr2_preds))
print("the MAPE of the test period using Gaussian Process Regressor with \nstandardized data (expanding window) is {}.".format(gpr2_mape))
```

## Q4

Summary of MAPE with different predict method: 

baseline(predict with same week in past years)                           : 0.043591387170382884<br>
Random Forest Regressor with previous 4 periods                          : 0.055117557801267245<br>
Random Forest Regressor with previous 4 periods and same period last year: 0.03619799944367144<br>
Gaussian Process Regressor with standardized data                        : 0.04486291984835997<br>
Gaussian Process Regressor with original data                            : 0.998106655270989<br>
Gaussian Process Regressor with standardized data(Expanding window)      : 0.04455348996054757.<br>

(1)The best one is Random Forest Regressor with previous 4 periods and same period last year and the worst one is Gaussian Process Regressor with original data. 

(2)The annual sales peak in December can not be predicted well in random forest regressor. 

(3)Predicting with the same period last year result in the best performance. It is probably because the seasonal influence is relative significant.
